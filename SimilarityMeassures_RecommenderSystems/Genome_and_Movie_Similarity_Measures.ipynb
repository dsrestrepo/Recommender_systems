# MSD WITOUT library
\begin{equation*}
msd(\mu ,\nu ) = \frac{1}{\left |I_{\mu ,\nu}  \right |}\cdot \sum_{i=0}^{I_{\mu ,\nu}}(r_{\mu i} - r_{\nu  i})^{2}
\end{equation*}
for users:
\begin{equation*}
\mu , \nu
\end{equation*}
with ratings:
\begin{equation*}
r_{\mu}, r_{\nu}
\end{equation*}
items in common:  
\begin{equation*}
I_{\mu ,\nu}
\end{equation*}

```python
#take the users:
ratingsUser1 = ratings[ratings.userId == 1]
ratingsUser1
```

```python
ratingsUser3 = ratings[ratings.userId == 3]
ratingsUser3
```

```python
def msd(user_u_ratings, user_v_ratings):
    i=0 #items in common
    difference=0 #array that saves the difference of each rating
    for movieu in user_u_ratings.movieId:
        for moviev in user_v_ratings.movieId:
            if movieu == moviev:
                i = i+1
                userUrating = user_u_ratings.rating[user_u_ratings.movieId==movieu].item()
                userVrating = user_v_ratings.rating[user_v_ratings.movieId==moviev].item()                          
                print(f'for user u the rating of {movieu} is {userUrating}')
                print(f'for user v the rating of {moviev} is {userVrating}')
                difference = difference + (userUrating - userVrating)**2
                print(f'the squared difference is {difference}')
    msd = difference/i
    print(f'the total of common movies rated is {i}')
    print(f'the sum of all ratings differences {difference}')
    print(f'the msd is {msd}')
    return msd
```

# we define our score:
    +4 if are the same
    0 if are different
    -2 if are different with an indel(-)

```python
def fpc(adn1,adn2):
    score=0
    indel=['A','C','T','G','-']
    if len(adn1) == len(adn2):
        #compare
        for i in range(0,len(adn1)):
            if adn1[i] == adn2[i]:
                score = score + 4
            else:
                if adn1[i] == '-' or adn2[i] == '-':
                    score = score -2
                else:
                    score = score + 0
        print(f'the score is {score}')
    elif len(adn1) < len(adn2):
        #add to adn1
        difference = len(adn2)-len(adn1)
        x=random.randint(len(indel), size=(difference))
        adding_value = ''
        for i in reversed(x):
            adding_value = adding_value + indel[i]
        adn1 = adn1+adding_value
        #compare
        for i in range(0,len(adn1)):
            if adn1[i] == adn2[i]:
                score = score + 4
            else:
                if adn1[i] == '-' or adn2[i] == '-':
                    score = score -2
                else:
                    score = score + 0
        print(f'the score is {score}')
        
    elif len(adn1) > len(adn2):
        #add to adn2
        difference = len(adn1)-len(adn2)
        x=random.randint(len(indel), size=(difference))
        adding_value = ''
        for i in reversed(x):
            adding_value = adding_value + indel[i]
        adn2 = adn2+adding_value
        #compare
        for i in range(0,len(adn1)):
            if adn1[i] == adn2[i]:
                score = score + 4
            else:
                if adn1[i] == '-' or adn2[i] == '-':
                    score = score -2
                else:
                    score = score + 0
        print(f'the score is {score}')
```
